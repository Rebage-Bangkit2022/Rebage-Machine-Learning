```python
import tensorflow as tf 
from tensorflow.keras.preprocessing import image
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
IMAGE_PATH = r'.\f1.jpeg'
INPUT_SIZE = 320
images_load = image.load_img(IMAGE_PATH, target_size=(INPUT_SIZE + 200, INPUT_SIZE))

x = image.img_to_array(images_load)
x = np.expand_dims(x, axis=0)
images = np.vstack([x])

images = images[0]
images = images/255.

plt.figure(figsize=(6, 12))
plt.imshow(images)
plt.show()
```

# Pengertian Brightness, Contrass, dan Saturation 
- Brightness (kecerahan) : Seberapa cerah suatu gambar
- Contrast (kontras) : Seberapa lebar perbedaan area terang dan gelap dalam sebuah gambar
- Saturation (kepekatan warna) : Seberapa pekat warna dalam gambar

## tf.image.random_brightness 

```python
images_brightness = tf.image.random_brightness(images, max_delta=.2, seed=101)

plt.figure(figsize=(6, 12))
plt.imshow(images_brightness)
plt.show()
```

## tf.image.random_contrast
"Contrast" (Kontras) merujuk ke perbedaan antara terang dan gelap. (semakin tinggi semakin gelap)

```python
images_contrass = tf.image.random_contrast(images, lower=.8, upper=1.1, seed=101)

plt.figure(figsize=(6, 12))
plt.imshow(images_contrass)
plt.show()
```

## random_saturation
"Saturation" (Saturasi) merujuk ke intensitas warna. (lebih jelas)

```python
images_saturation = tf.image.random_saturation(images, lower=.8, upper=1.1, seed=101)

plt.figure(figsize=(6, 12))
plt.imshow(images_saturation)
plt.show()
```
