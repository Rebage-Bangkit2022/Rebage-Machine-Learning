```python
import os
import cv2
import numpy as np 
import matplotlib.pyplot as plt
```

```python
def showImageFromNumpy(image_name):
    arr = np.load(image_name)
    # plt.figure(figsize=(12, 8))
    plt.imshow(cv2.cvtColor(arr, cv2.COLOR_BGR2RGB))
    plt.show()
```

```python
showImageFromNumpy('gambar1.npy')
```
